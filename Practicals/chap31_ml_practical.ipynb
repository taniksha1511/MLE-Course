```python
import sys 
import numpy as np 
import pandas as pd
import matplotlib
import sklearn

print("=== Library Versions ===")
print(f"Python    : {sys.version.split()[0]}")
print(f"NumPy     : {np.__version__}")
print(f"Pandas    : {pd.__version__}")
print(f"Matplotlib: {matplotlib.__version__}")
print(f"scikit-Learn   : {sklearn.__version__}")
```

```python
from sklearn.linear_model import LinearRegression
model= LinearRegression()
x=[[1],[2],[3],[4]]
y=[2,4,6,8]
model.fit(x,y)
pred=model.predict([[6]])
print("Prediction for input 6 is: ",pred)
```

**User input**

```python
from sklearn.linear_model import LinearRegression
model= LinearRegression()
x=[[1],[2],[3],[4]]
y=[2,4,6,8]
model.fit(x,y)
user_value=float(input("Enter a number to predict: "))
pred=model.predict([[user_value]])
print("Prediction for input 6 is: ",pred[0])
```

```python
import matplotlib.pyplot as plt
plt.scatter(x,y, color='blue', label='Data Points')
plt.plot(x,model.predict(x), color='orange', label="Regression Line")
plt.legend(loc='center left',bbox_to_anchor=(1,0.5))
plt.title("Linear Regression")
plt.xlabel("X")
plt.xlabel("Y")
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
model= LinearRegression()
weights=np.array([1,2,3,4,5]).reshape(-1,1)
prices=np.array([30,60,90,120,150])
model.fit(weights,prices)
user_value=float(input("Enter a number to predict: "))
pred=model.predict([[user_value]])
print("Prediction for input 6 is: ",pred[0])
plt.scatter(weights,prices, color='blue', label='Data Points')
plt.plot(weights,model.predict(weights), color='orange', label="Regression Line")
plt.legend(loc='center left',bbox_to_anchor=(1,0.5))
plt.title("Linear Regression")
plt.xlabel("weights")
plt.xlabel("prices")
plt.tight_layout()
plt.show()
```
